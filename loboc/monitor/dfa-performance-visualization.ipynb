```python
import numpy as np
import matplotlib.pyplot as plt

from dateutil import parser

# input dataset
fileName = 'performance-dt=0.100'
data = np.genfromtxt(fileName + '.log', dtype=None, delimiter=';', names=True) 
# format timestamp column
datetimeParser = lambda t: parser.parse(t.decode("utf-8"))
timestamps = np.vectorize(datetimeParser)(data['timestamp'])
print(data.dtype.names)
# print(data)
```

```python
# plots
x = timestamps
y1 = data['transactions_generated']
y2 = data['transactions_processed']
plt.plot(x, y1, 'red')
plt.plot(x, y2, 'blue')
plt.savefig(fileName + '.png')
plt.show()
```
